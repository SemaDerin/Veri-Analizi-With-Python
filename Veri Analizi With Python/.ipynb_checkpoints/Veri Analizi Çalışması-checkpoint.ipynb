```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('Kitap1.csv',encoding='cp1254',header=None, names=["Computer", "Price", "Rate","Kapasite","RAM","İşlemci Tipi","Ekran Boyutu","İşletim Sistemi","Seller","Brand"])

```

Önce gerekli kütüphaneleri import ettik ve verisetini içe aktardık ve başlıkları belirledik.

Veri setimizin ilk beş satırını görüntülemek ile işe başladık.

```python
df.head()
```

Marka sütunun içeriğini tamamen büyük harfler yapıp ardından yeni veri seti olarak kaydettik.

```python
df['Brand'] = df['Brand'].str.upper()
df.to_csv('yeni_veriseti.csv', index=False)
```

Burada yapmayı hedeflediğimiz işlem, her bir markanın fiyatlarına göre bir histogram grafiği çizdirdik.

Grafiği analiz edersek 0-10000TL arasında frekans oldukça fazla. Yani verisetimizin genellikle 0-10000 tl arasında ürünlerden(Bilgisayar) meydana geliyor.

```python
df.groupby('Brand')['Price'].plot(kind='hist', alpha=0.5, legend=True, figsize=(10,6))
plt.title('Farklı Markalardaki Bilgisayarların Fiyat Dağılımı')
plt.xlabel('Price')
plt.ylabel('Frekans')
plt.show()
```

Veri setinizdeki fiyat verilerini belirli aralıklara bölüp ve her aralıkta hangi markaların tercih edildiğine dair bir analiz yapmak için Pandas kütüphanesinde yer alan pd.cut() fonksiyonunu kullandık. Bu fonksiyon, veri setindeki bir sütunun değerlerini belirli aralıklara böler ve bu aralıkları yeni bir sütun olarak veri setine ekler.

Aşağıdaki kodda, "Fiyat" sütunundaki veriler 3 farklı aralığa bölünerek "Fiyat Aralığı" adlı yeni bir sütuna eklenir. Daha sonra bu veri seti üzerinde, her fiyat aralığındaki markaların sayısını sayarak hangi fiyat aralığında hangi markaların tercih edildiğine dair bir tablo oluşturduk.

```python
df['Fiyat Aralığı'] = pd.cut(df['Price'], bins=[0, 5000, 10000, 15000], labels=['0-5000', '5001-10000', '10001-15000'])

# Her fiyat aralığındaki markaların sayısını sayma
fiyat_marka = pd.crosstab(df['Fiyat Aralığı'], df['Price'])

# Sonucu ekrana yazdırma
print(fiyat_marka)
```

```python
df['Fiyat Aralığı'] = pd.cut(df['Price'], bins=[0, 5000, 10000, 15000], labels=['0-5000', '5001-10000', '10001-15000'])

# Her fiyat aralığındaki markaların sayısını sayma
fiyat_marka = pd.crosstab(df['Fiyat Aralığı'], df['Price'])
fiyat_marka.plot(kind='bar', stacked=True)
plt.title('Fiyat Aralığına Göre Marka Tercihleri')
plt.xlabel('Fiyat Aralığı')
plt.ylabel('Marka Sayısı')
plt.show()
```
